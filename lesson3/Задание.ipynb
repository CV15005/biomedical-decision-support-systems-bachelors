```python
# Загрузка медицинского датасета для сравнения классификаторов

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Загружаем датасет рака груди (бинарная медицинская классификация)
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names
class_names = data.target_names

# Разбиение на train/test с сохранением пропорций классов
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Стандартизация признаков (важно для многих моделей: логистическая регрессия, SVM, kNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Train shape:", X_train.shape)
print("Test shape: ", X_test.shape)
print("Классы:", class_names)

# Дальше во всех подзаданиях этого блокнота
# используйте X_train/X_test или X_train_scaled/X_test_scaled
```
